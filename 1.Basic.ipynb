<h1> Tangent vs Slope <h1>

A tangent to a curve is a line that touches the curve at a single point, but does not intersect the curve at any other points. The slope of a tangent 
to a curve at a particular point is equal to the derivative of the curve at that point.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define a function
def f(x):
  return x**2 + 3*x + 2

# Define the x-values for the function
x = np.linspace(-2, 3, 50)

# Calculate the y-values for the function
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x)')

# Calculate and plot the derivative of the function
def f_derivative(x):
  return 2*x + 3

y_derivative = f_derivative(x)
plt.plot(x, y_derivative, label="f'(x)")

# Add a legend and axis labels
plt.legend(loc='upper left')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```
