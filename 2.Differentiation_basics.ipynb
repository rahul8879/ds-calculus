<h3> Example </h3>

Suppose you have a function f(t) that represents the height of a bouncing ball as a function of time. The function might take the form:

f(t) = -4.9t^2 + vt + h

Where t is the time (in seconds), v is the initial velocity of the ball (in meters per second), and h is the initial height of the ball (in meters).

To find the velocity of the ball at each moment in time, you can take the derivative of the function with respect to time:

df/dt = -9.8t + v


This derivative represents the rate at which the height of the ball is changing over time, which is also known as the velocity of the ball.

Now, suppose you want to determine whether the ball is moving up or down at a given time t. You can do this by evaluating the derivative at that time:

velocity = -9.8t + v

If the velocity is positive, the ball is moving upwards. If the velocity is negative, the ball is moving downwards.


```python
import numpy as np
import matplotlib.pyplot as plt

# Set the initial velocity and height of the ball
v =10
h = 10

# Create a range of time values to plot
t = np.linspace(0, 1, 1000)

# Calculate the height and velocity of the ball at each time
y = -4.9*t**2 + v*t + h
dy = -9.8*t + v

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1)

# Plot the height of the ball in the top subplot
ax1.plot(t, y)
ax1.set_ylabel('Height (m)')

# Plot the velocity of the ball in the bottom subplot
ax2.plot(t, dy)
ax2.set_ylabel('Velocity (m/s)')

# Show the plot
plt.show()
```

<h3> Application of Differentiation in ML </h3>

In the field of machine learning, differentiation is used in a number of ways. Some examples include:

<ul>
<li>Optimizing a model's parameters: In many machine learning algorithms, we need to find the values of the model's parameters that minimize some loss function. Differentiation can be used to find the gradient of the loss function with respect to the model's parameters, which can then be used to update the parameters in an iterative optimization process.</li>
<li>Backpropagation: In neural networks, the process of training a model involves adjusting the weights and biases of the network to minimize the loss function. This is done using an algorithm called backpropagation, which relies on differentiation to calculate the gradients of the loss function with respect to the network's weights and biases.</li>
<li>Autodiff: Many machine learning libraries, such as TensorFlow and PyTorch, provide support for automatic differentiation (autodiff), which allows you to define a computational graph and automatically compute the gradients of a function with respect to its inputs. Autodiff can be used to simplify the implementation of machine learning algorithms and to make it easier to optimize model parameters</li></ul>

Overall, differentiation is an important concept in machine learning and is used in many different contexts to optimize models and improve their performance.

<h3> Differentiation using 1st Principle </h3>

```python
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return x**2

# Define the derivative function
def derivative(f, x, h=1e-9):
    return (f(x+h) - f(x)) / h

# Define the point x0 at which to draw the tangent line
x0 = 0.5

# Calculate the derivative of f(x) at x0
df = derivative(f, x0)

# Calculate the y-intercept of the tangent line
b = f(x0) - df * x0

# Define a range of x values to plot
x_vals = np.linspace(-2, 2, 100)

# Calculate the derivative of f(x) at each x value using the derivative function
y_vals = [derivative(f, x_val) for x_val in x_vals]

# Calculate the y-values of the tangent line
tangent_y_vals = df * x_vals + b

# Plot the function and the tangent line
plt.plot(x_vals, f(x_vals), label='f(x)')
plt.plot(x_vals, tangent_y_vals, label='Tangent line')

# Add a legend and label the axes
plt.legend()

plt.show()
```
